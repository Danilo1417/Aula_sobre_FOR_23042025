```python
lista_vendas = [3000,5000,4000,1500,2000,2300]
meta = 1200
percentual_bonus = 0.50
for venda in lista_vendas:
  if venda >= meta:
    bonus = percentual_bonus * venda
  else:
    bonus = 0
  print(bonus)
```

```python
vendas_diarias = [20,30,40,50,60,70,80,90,100]
meta = 30
dias_acima_da_meta = 0

for diaria in vendas_diarias:
    if diaria > meta:
      #dias_acima_da_meta = dias_acima_da_meta + 1
      dias_acima_da_meta += 1      #atribuição

print(dias_acima_da_meta)
```

```python
vendedores = [10,11,22,15,30,10,14,21,23,55,21,60]
meta = 60
dias_acima_da_meta = 0
for quantidades_de_vendas in vendedores:
   if quantidades_de_vendas <= meta:
      dias_acima_da_meta += 1
print(dias_acima_da_meta)
```
